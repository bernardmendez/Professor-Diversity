```python
two = pd.read_csv('Dem-By-School copy.csv')
two = two.fillna(0)
two.head()
```

```python
dept_names = []
data = []
for a in range (int(len(two)/10)):
    x = a*10
    dept_names.append(two.iloc[x][0])
    temp = []
    for b in range (9):
        temp.append(two.iloc[x+b+1])
    data.append(temp)
dept_names
data
```

```python
temp2 = pd.DataFrame(data[0])#.drop(columns=['Name'])
#.drop(columns=['Name'])
for a in range (10):
    temp2 = temp2.rename(index={a: data[0][a-1]['Name']})
temp2
```

```python
temp3 = []
for i in range (len(data)):
    temp2 = pd.DataFrame(data[i])#.drop(columns=['Name'])
    for a in range (10):
        temp2 = temp2.rename(index={a: data[i][a-1]['Name']})
    temp3.append(temp2)
temp3[0]['Female']
```

```python
years = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018]
```

```python
for j in range (len(dept_names)):
    print(dept_names[j])
    fem = temp3[j]['Female']
    mal = temp3[j]['Male']
    fem_perc = []
    tota = []
    for k in range (9):
        tot = float(mal.iloc[k]) + float(fem.iloc[k])
        tota.append(tot)
        fem_p = float(fem.iloc[k]) / tot
        fem_perc.append(fem_p)
    fem_perc = np.array(fem_perc)
    tota = np.array(tota)
    m, b = np.polyfit(tota, fem_perc, 1)
    print(m)
    print(b)
    plt.plot(tota, fem_perc, 'o')
    plt.plot(tota, m * fem_perc + b)
    plt.show()
```
